5. Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
# Создаем матрицу A в шахматном порядке
matrix_A = np.array([[5 if (i + j) % 2 == 0 else 0 for j in range(5)] for i in range(5)])

# Создаем матрицу B с треугольной формой
matrix_B = np.tri(5, 5, k=0) * 5  # Создаем верхнюю треугольную матрицу и умножаем на 5

# Вычисляем детерминанты
determinant_A = np.linalg.det(matrix_A)
determinant_B = np.linalg.det(matrix_B)

# Вычисляем обратные матрицы (если возможно)
try:
    inverse_A = np.linalg.inv(matrix_A)
except np.linalg.LinAlgError:
    inverse_A = None

try:
    inverse_B = np.linalg.inv(matrix_B)
except np.linalg.LinAlgError:
    inverse_B = None

# Выводим результаты
print("Матрица A:")
print(matrix_A)
print("Детерминант матрицы A:", determinant_A)
print("Обратная матрица A:")
print(inverse_A)

print("\nМатрица B:")
print(matrix_B)
print("Детерминант матрицы B:", determinant_B)
print("Обратная матрица B:")
print(inverse_B)
```
