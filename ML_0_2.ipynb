```python
data = data1['Задания с кратким ответом'].to_list()
k = len(data[0])
correct_answers = [0] * k
wrong_answers = [0] * k
skipped_answers = [0] * k

for line in data:
    for i, char in enumerate(line):
        if char == "+":
            correct_answers[i] += 1
        elif char == "-":
            wrong_answers[i] += 1
        elif char.isdigit():
            skipped_answers[i] += 1

total_students = len(data)
correct_percentages = [(x / total_students) * 100 for x in correct_answers]
wrong_percentages = [(x / total_students) * 100 for x in wrong_answers]
skipped_percentages = [(x / total_students) * 100 for x in skipped_answers]
```
