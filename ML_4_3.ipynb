# 1


```python
from sklearn.datasets import load_iris

# Загрузка датасета
iris = load_iris()

# Просмотр информации о датасете
print("Keys of iris dataset:", iris.keys())
print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)
print("Number of samples:", len(iris.data))
print("Shape of data:", iris.data.shape)
```

# 2

```python
from sklearn.model_selection import train_test_split

# Разделение данных на обучающую и валидационную выборки
X_train, X_val, y_train, y_val = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Вывод индексов разбиения
print("Индексы обучающей выборки:", list(range(len(X_train))))
print("Индексы валидационной выборки:", list(range(len(X_train), len(X_train) + len(X_val))))
```

# 3

```python
import numpy as np

# Задаем seed для воспроизводимости
np.random.seed(42)

# Создаем случайную перестановку индексов
permuted_indices = np.random.permutation(len(iris.data))

# Применяем перестановку к данным и меткам
X_permuted = iris.data[permuted_indices]
y_permuted = iris.target[permuted_indices]

# Разделение перемешанных данных на обучающую и валидационную выборки
X_train_permuted, X_val_permuted, y_train_permuted, y_val_permuted = train_test_split(X_permuted, y_permuted, test_size=0.2, random_state=42)

# Вывод индексов разбиения
print("Индексы обучающей выборки (перемешанные):", list(permuted_indices[:len(X_train_permuted)]))
print("Индексы валидационной выборки (перемешанные):", list(permuted_indices[len(X_train_permuted):]))
```

# 4

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

# Обучение модели логистической регрессии
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Вывод значений коэффициентов модели
print("Значения коэффициентов модели:", log_reg.coef_)

# Предсказание на тестовом наборе признаков
y_pred = log_reg.predict(X_val)

# Вычисление и вывод метрик accuracy и f1-score
accuracy = accuracy_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred, average='weighted')
print("Accuracy:", accuracy)
print("F1-score:", f1)
```

# 5

```python
# Разделение данных на обучающую и валидационную выборки (новое соотношение 75-25)
X_train_new, X_val_new, y_train_new, y_val_new = train_test_split(iris.data, iris.target, test_size=0.25, random_state=42)

# Обучение модели логистической регрессии на новых данных
log_reg_new = LogisticRegression()
log_reg_new.fit(X_train_new, y_train_new)

# Вывод значений коэффициентов модели для новых данных
print("Значения коэффициентов модели для новых данных:", log_reg_new.coef_)

# Предсказание на валидационном наборе признаков для новых данных
y_pred_new = log_reg_new.predict(X_val_new)

# Вычисление и вывод метрик accuracy и f1-score для новых данных
accuracy_new = accuracy_score(y_val_new, y_pred_new)
f1_new = f1_score(y_val_new, y_pred_new, average='weighted')
print("Accuracy для новых данных:", accuracy_new)
print("F1-score для новых данных:", f1_new)

# Сравнение метрик с предыдущими значениями
print("\nСравнение с предыдущими значениями:")
print("Разница в accuracy:", accuracy_new - accuracy)
print("Разница в F1-score:", f1_new - f1)
```

```python
from sklearn.model_selection import cross_val_score

# Выполнение кросс-валидации модели логистической регрессии
scores = cross_val_score(LogisticRegression(), iris.data, iris.target, cv=5)

# Вывод результатов кросс-валидации
print("Значения метрик accuracy для каждого блока:", scores)
print("Среднее значение метрики accuracy по всем блокам:", scores.mean())
```

```python
from sklearn.model_selection import cross_val_score

# Выполнение кросс-валидации модели логистической регрессии с помощью cross_val_score
scores = cross_val_score(LogisticRegression(), iris.data, iris.target, cv=5)

# Вывод результатов кросс-валидации
print("Значения метрик accuracy для каждого блока:", scores)
print("Среднее значение метрики accuracy по всем блокам:", scores.mean())
```

# 8

```python
from sklearn.model_selection import StratifiedKFold

# Создаем объект StratifiedKFold для кросс-валидации с 5 блоками
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Создаем модель логистической регрессии
log_reg = LogisticRegression()

# Список для хранения значений accuracy для каждого блока
scores_stratified = []

# Цикл по блокам кросс-валидации
for train_index, test_index in stratified_kfold.split(iris.data, iris.target):
    X_train_stratified, X_test_stratified = iris.data[train_index], iris.data[test_index]
    y_train_stratified, y_test_stratified = iris.target[train_index], iris.target[test_index]
    
    # Обучение модели
    log_reg.fit(X_train_stratified, y_train_stratified)
    
    # Оценка точности на тестовом наборе
    accuracy_stratified = log_reg.score(X_test_stratified, y_test_stratified)
    scores_stratified.append(accuracy_stratified)

# Вывод результатов кросс-валидации с стратификацией
print("Значения метрик accuracy для каждого блока (со стратификацией):", scores_stratified)
print("Среднее значение метрики accuracy по всем блокам (со стратификацией):", np.mean(scores_stratified))
```

# 9

```python
from sklearn.model_selection import LeaveOneOut

# Создаем объект LeaveOneOut для кросс-валидации
leave_one_out = LeaveOneOut()

# Создаем модель логистической регрессии
log_reg = LogisticRegression()

# Список для хранения значений accuracy для каждого блока
scores_leave_one_out = []

# Цикл по блокам кросс-валидации
for train_index, test_index in leave_one_out.split(iris.data):
    X_train_loo, X_test_loo = iris.data[train_index], iris.data[test_index]
    y_train_loo, y_test_loo = iris.target[train_index], iris.target[test_index]
    
    # Обучение модели
    log_reg.fit(X_train_loo, y_train_loo)
    
    # Оценка точности на тестовом наборе
    accuracy_leave_one_out = log_reg.score(X_test_loo, y_test_loo)
    scores_leave_one_out.append(accuracy_leave_one_out)

# Вывод результатов кросс-валидации с leave-one-out разбиением
print("Значения метрик accuracy для каждого блока (leave-one-out):", scores_leave_one_out)
print("Среднее значение метрики accuracy по всем блокам (leave-one-out):", np.mean(scores_leave_one_out))
```
